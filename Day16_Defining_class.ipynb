```python
#REQ:create a greeting class.....

1.it needs to accept input from the user name-->method1
2.it needs to display the name--->method2
3.it needs to greet the name-->method3
```

```python
fun--->def--->small case
class--->title case----->class is the keyword we have to use
```

```python
class greeting:   #declaring the class
    """creating a greeting class for the req"""   #Doc string
    def createname(self,name):  #Method1
        self.name=name        #attribute(var)--->x.obj.name='shafi'
    def display(self):          #Method2
        print(self.name)      
    def greet(self):           #Method3
        print(f"welcome,{self.name}")
```

```python
#obj creation and assigning it to class...!
```

```python
xobj=greeting()  #obj has beeen assigned to class
```

```python
xobj.createname('shafi') #obj.method
```

```python
xobj.display()
```

```python
xobj.greet()
```
