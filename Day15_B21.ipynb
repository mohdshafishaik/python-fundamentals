```python
def describre_pet(animal_type,animal_name):
    """Display information about a pet"""
    print(f"\nI have a {animal_type}")
    print(f"my {animal_type}'s name is {animal_name}")
```

```python
describre_pet('dog','whiskers')
```

```python
describre_pet('broody','dog')  #-->this is wrong
```

```python
#keyword arguments:
```

```python
describre_pet(animal_name='broody',animal_type='dog')
```

```python
describre_pet(animal_type='dog','broody')
```
