```python
prompt='\nTell me something and i will repeat it for you'
prompt+='\n Enter "quit" to quit'
```

```python
message=''
while message!="quit": #not equal to
    message=input(prompt)
    print(message)
```

```python
prompt='\nPlease enter a city name you have visited:'
prompt+='\nEnter "quit" when you are finished:'
```

```python
city=''
while True:
    city=input(prompt)
    
    if city=="quit":
        break  #It will be stopping the execution of the program...
    else:
        print(f"\n I would love to go to the {city.title()}.")
```

```python
continue means it will be validating till the end of the program.
```

```python
current_num=0

while current_num<10:
    current_num+=1
    if current_num%2==0: #even condition
        continue
    
    print(current_num)
```
