```python
class dog:
    """creating a dog class and defining the methods of the dog"""
    
    def __init__(self,name,age):
        self.name=name
        self.age=age
        print("this line got printed automatically")
    def sit(self):
        print(f"{self.name} is now sitting on the command")
    def jump(self):
        print(f"{self.name} is now jumping on command")
    def roll_over(self):
        print(f"{self.name} is now rolling over")
    def sleep(self):
        print(f"{self.name} is now sleeping")
    def details(self):
        print(f"the dog's name is {self.name}")
        print(f"the dog's age is {self.age}")
```

```python
x=dog('shiro',4)
```

```python
x.sit()
```

```python
x.jump()
```

```python
x.roll_over()
```

```python
x.sleep()
```
